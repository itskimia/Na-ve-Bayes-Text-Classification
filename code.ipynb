```python
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))
nltk.download('wordnet')
import string
import re
string.punctuation
```

```python
#importing the stopwords from the file given
sw = pd.read_csv(r'C:\Users\hashemi\Desktop\Term6\داده کاوی\HW3\sw.txt')
stopwords = sw['about'].to_list()
stopwords.append('about')


training_data['Review'] = training_data['Review'].map(lambda x: re.sub('[,\.!?]', '', x))


#lowercase
training_data['Review'] = training_data['Review'].map(lambda x: x.lower())
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

#removing punctuations
training_data['Review'] = training_data['Review'].apply(lambda x: remove_punct(x))

def tokenization(text):
    text = re.split('\W+', text)
    return text
#tokenizing
training_data['Review'] = training_data['Review'].apply(lambda x: tokenization(x.lower()))

def remove_stopwords(text):
    text = [word for word in text if word not in stopwords]
    return text
#removing the stopwords
training_data['Review'] = training_data['Review'].apply(lambda x: remove_stopwords(x))
```

```python
#importing testing data and cleaning them
sw = pd.read_csv(r'C:\Users\hashemi\Desktop\Term6\داده کاوی\HW3\sw.txt')
stopwords = sw['about'].to_list()
stopwords.append('about')

testing_data['Review'] = testing_data['Review'].map(lambda x: re.sub('[,\.!?]', '', x))

testing_data['Review'] = testing_data['Review'].map(lambda x: x.lower())
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

testing_data['Review'] = testing_data['Review'].apply(lambda x: remove_punct(x))

def tokenization(text):
    text = re.split('\W+', text)
    return text

testing_data['Review'] = testing_data['Review'].apply(lambda x: tokenization(x.lower()))

def remove_stopwords(text):
    text = [word for word in text if word not in stopwords]
    return text
    
testing_data['Review'] = testing_data['Review'].apply(lambda x: remove_stopwords(x))
```

```python
data0 = training_data[training_data['Label'] == 'neg']
data1 = training_data[training_data['Label'] == 'pos']
numofP = len(data0)
numofN = len(data1)
probPos = float(numofP)/float(len(training_data))
probNeg = float(numofN)/float(len(training_data))
```

```python
#predicting the testing_data
from math import log2

testing_data['Result'] = 0*len(testing_data)
for i in range(len(testing_data)):
    prob1 = probPos
    prob2 = probNeg
    for j in range(len(testing_data.at[i,'Review'])):
        k = (testing_data.at[i,'Review'])[j] #words of each sentence
        
        if(k in dict1.keys()):
            prob1 = prob1 * log2(dict1[k])
        if(k in dict2.keys()):
            prob2 = prob2 * log2(dict2[k])

    if(abs(prob1) > abs(prob2)):
        testing_data.at[i,'Result'] = 1
```
